```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv('ool_pds.csv', low_memory=False)
```

```python
data.head()
```

```python
#converting to numeric
data['PPINCIMP'] = pd.to_numeric(data['PPINCIMP'], errors='coerce')
data['W2_QE1'] = pd.to_numeric(data['W2_QE1'], errors='coerce')
#replacing category labels with 0 and 1 for simplicity
data['W2_QE1'] = data['W2_QE1'].replace(to_replace={1: 0, 2: 1})
#excluding missing and incorrect values
data = data.dropna(subset=['PPINCIMP', 'W2_QE1'])
data = data.query('W2_QE1 in [0, 1]')
#subtracting the mean from explanatory variable
data['PPINCIMP'] = data['PPINCIMP'] - data['PPINCIMP'].mean()
data.shape
```

```python
logr = smf.logit(formula='W2_QE1~PPINCIMP', data=data)
result = logr.fit()
result.summary()
```
